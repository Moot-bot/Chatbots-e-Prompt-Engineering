```python

# =============================================
# üß† Configura√ß√£o do LLM
# =============================================
os.environ["OPENAI_API_KEY"] = "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"  # Substitua pela sua chave
llm = OpenAI(model="gpt-4o-mini")

# =============================================
# üß© Fun√ß√£o auxiliar para formatar respostas
# =============================================
def format_text(response):
    text = response.response if hasattr(response, "response") else str(response)
    print(textwrap.fill(text, width=100))


# =============================================
# üîó Fun√ß√£o principal: pipeline de consulta
# =============================================
def pipeline_consulta(df):
    """Cria um pipeline que transforma perguntas em consultas pandas e respostas interpretadas."""

    # --- Instru√ß√µes ao modelo ---
    instruction_str = (
        "1. Converta a consulta para c√≥digo execut√°vel usando pandas.\n"
        "2. A linha final do c√≥digo deve ser uma express√£o Python execut√°vel via eval().\n"
        "3. O c√≥digo deve representar a solu√ß√£o da consulta.\n"
        "4. Imprima apenas a express√£o.\n"
        "5. N√£o coloque a express√£o entre aspas.\n"
    )

    pandas_prompt_str = (
        "Voc√™ est√° trabalhando com um DataFrame do pandas chamado `df`.\n"
        "Este √© o resultado de `print(df.head())`:\n"
        "{df_str}\n\n"
        "Siga estas instru√ß√µes:\n"
        f"{instruction_str}\n\n"
        "Consulta: {{query_str}}\n\n"
        "Express√£o:"
    )

    response_synthesis_prompt_str = (
        "Dada uma pergunta de entrada, atue como analista de dados e elabore uma resposta natural "
        "com base no resultado da consulta pandas.\n\n"
        "Consulta: {query_str}\n"
        "C√≥digo gerado: {pandas_instructions}\n\n"
        "Sa√≠da do Pandas: {pandas_output}\n\n"
        "Resposta:"
        "Ao final, exiba o c√≥digo usado: O c√≥digo utilizado foi {pandas_instructions}"
    )

    # --- Cria√ß√£o dos templates ---
    pandas_prompt = PromptTemplate(pandas_prompt_str).partial_format(df_str=df.head())
    pandas_output_parser = PandasInstructionParser(df)
    response_synthesis_prompt = PromptTemplate(response_synthesis_prompt_str)

    # --- Cria√ß√£o do pipeline ---
    qp = QP(
        modules={
            "input": InputComponent(),
            "pandas_prompt": pandas_prompt,
            "llm1": llm,
            "pandas_output_parser": pandas_output_parser,
            "response_synthesis_prompt": response_synthesis_prompt,
            "llm2": llm,
        },
        verbose=False,
    )

    qp.add_chain(["input", "pandas_prompt", "llm1", "pandas_output_parser"])
    qp.add_links([
        Link("input", "response_synthesis_prompt", dest_key="query_str"),
        Link("llm1", "response_synthesis_prompt", dest_key="pandas_instructions"),
        Link("pandas_output_parser", "response_synthesis_prompt", dest_key="pandas_output"),
    ])
    qp.add_link("response_synthesis_prompt", "llm2")

    return qp


# =============================================
# üíæ Fun√ß√µes auxiliares do Gradio
# =============================================
def load_data(file_path, df_state):
    """Carrega um CSV enviado e retorna um DataFrame."""
    if file_path is None or file_path == "":
        return "Fa√ßa o upload de um dataset para come√ßar a an√°lise.", df_state
    try:
        df = pd.read_csv(file_path)
        return "‚úÖ Arquivo carregado com sucesso!", df
    except Exception as e:
        return f"‚ùå Erro ao carregar o dataset: {str(e)}", df_state


def process_question(question, df_state):
    """Executa o pipeline de an√°lise baseado na pergunta."""
    if df_state is not None and question:
        qp = pipeline_consulta(df_state)
        response = qp.run(query_str=question)
        return response.message.content if hasattr(response, "message") else str(response)
    return "Por favor, carregue um dataset primeiro e fa√ßa uma pergunta."


# =============================================
# üß∞ Interface Gradio
# =============================================
with gr.Blocks() as app:
    gr.Markdown("# üìä Plataforma de An√°lise Autom√°tica de Dados com LlamaIndex")

    with gr.Row():
        in_file = gr.File(label="üìÇ Upload CSV", file_count="single", type="filepath")

    with gr.Row():
        in_question = gr.Textbox(label="‚ùì Pergunte algo sobre os dados", placeholder="Ex: Qual o tipo de opera√ß√£o mais comum por ano?")
        bt_submit = gr.Button("Enviar")

    with gr.Row():
        out_answer = gr.Textbox(label="üß† Resposta", lines=6)

    df_state = gr.State(value=None)

    in_file.change(fn=load_data, inputs=[in_file, df_state], outputs=[out_answer, df_state])
    bt_submit.click(fn=process_question, inputs=[in_question, df_state], outputs=out_answer)

# =============================================
# ‚ñ∂Ô∏è Executar App
# =============================================
app.launch(debug=True)
```
